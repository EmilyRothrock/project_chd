```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('fhs_train.csv')
data = data.drop('Unnamed: 0',axis=1)
data
```

```python
columns = ['sex', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD']
```

```python
# RESOURCES
# https://www.statology.org/pandas-select-rows-without-nan/

data = data[~data.isnull().any(axis=1)] # gets rid of all rows with np.nan values
data = data.rename(columns={"TenYearCHD": "TEN_YEAR_CHD"})
data.to_csv("fhs_test_cleaned.csv")
data
```

```python
print("Highest correlations come from age, prevalentHyp, sysBP, glucose, diaBP and totChol with the highest being age and sysBP")
```

```python
correlated_columns = ['age', 'prevalentHyp', 'sysBP', 'glucose', 'diaBP', 'totChol']
for column in correlated_columns:
    print(data.loc[:,[column, 'TEN_YEAR_CHD']].groupby(column).describe())
```
